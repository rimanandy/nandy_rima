**Percentage of People having heart disease**

```python
target = df.groupby("target").size()
target
```

```python
#Converting numerical data into categorical data
def heart_disease(row):
  if row ==0:
    return 'Absence'
  elif row==1:
    return "Presence"
```

```python
df["Heart_disease"]=df['target'].apply(heart_disease)
df.head()
```

```python
heart_disease_count=df.groupby('Heart_disease')["target"].count()
heart_disease_count
```

```python
#Pie chart creation of Heart disease population
plt.figure(figsize=(10,6))
plt.pie(heart_disease_count,labels=["Absence","Presence"],autopct='%0.0f%%')
plt.title('Heart Disease Population Percentage')
plt.show()

```

People who have heart disease is 51% whereas people who donot have heart disease is 49%. Thus, the presence of heart disease is 2% higher than its absence.

```python
#Countplot creation of population vs age
plt.figure(figsize=(10,6))
sns.countplot(x="age",data=df)
plt.title("Population Count")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

In this section,the best analysis can be divided into young age, middle age and Elderly ages.

```python
#Statistical analysis
Min_age= df["age"].min()
Max_age=df["age"].max()
Mean_age=df["age"].mean()
print("The minimum age is:",Min_age)
print('The maximum age is:',Max_age)
print('The mean age is:',Mean_age)
```

```python
#Categorical Analysis
Young_ages=df[(df["age"]>=29)&(df["age"]<40)]
Middle_ages=df[(df["age"]>=40)&(df["age"]<55)]
Elderly_ages=df[(df["age"]>55)]
print("Young ages=",len(Young_ages))
print("Middle ages=",len(Middle_ages))
print("Elderly ages=",len(Elderly_ages))
```

```python
#Barplot using age category
sns.barplot(x=["Young_ages","Middle_ages","Elderly_ages"],y=[len("Young_ages"),len("Middle_ages"),len("Elderly_ages")],palette='YlGn_r')
plt.title("Age Category")
plt.xlabel("Age Range")
plt.ylabel("Total Count")
plt.show()

```

```python
#Converting numerical data into categorical data
def gender(row):
  if row ==1:
    return 'Male'
  else:
    return "Female"
```

```python
df["sex1"]=df['sex'].apply(gender)
df
```

```python
def age_range(row):
  if row>=29 and row<40:
    return "Young age"
  elif row>=40 and row<=55:
    return "Middle age"
  else:
    return "Elderly age"
```

```python
df["Age Range"]=df["age"].apply(age_range)
df
```

```python
#Count Plot creation using Age Category
plt.figure(figsize=(10,7))
sns.countplot(x='Age Range',data=df,hue='sex1')
plt.title('Gender Based Age Category')
plt.xlabel('Age Category')
plt.ylabel('Age')
plt.show()
```

In our population the number of males and females are both maximum in the elderly age category.

```python
#Count Plot Creation of Heart Disease Based On Age Category using MatplotLib and Seaborn
plt.figure(figsize=(7,5))
sns.countplot(x=df['Heart_disease'],data=df,hue='Age Range',palette="Pastel1")
plt.title(' Heart Disease Based on Age')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.show()

```

## From the above count plot, it can be clearly inferred that the middle aged people are prone to heart disease than the elderly and the young aged.

```python
#Count Plot Creation of Heart Disease Based on Gender using MatplotLib and Seaborn

plt.figure(figsize=(7,5))
sns.countplot(x=df['Heart_disease'], hue='sex1', data=df, palette='BuGn_r')
plt.xlabel('Heart Disease')
plt.ylabel('Count')
plt.legend(labels=['Male','Female'])
plt.title('Heart Disease Based on Gender')
plt.show()
```

We can see that females are more prone to heart disease than males.

```python
# count plot for chest pain experienced in our population
sns.countplot(x=df["Heart_disease"],hue="cp",data=df,order=['Presence',"Absence"])
plt.title("Chest Pain Experienced")
plt.xlabel("Heart Disease")
plt.ylabel("Count")
plt.legend(labels=['Typical Angina','Atypical Angina','Non-Anginal pain','Asymptomatic'])
plt.show()
```

It seems people who have non-anginal pain have a higher chance of heart disease.
